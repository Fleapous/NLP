```python
import pandas as pd
import numpy as np
```

```python
ft_pd = pd.read_csv("revised_annotations_as_of_May31st.csv", sep=';')
ft_pd.fillna('NA', inplace=True)
res_pd = pd.read_csv("processed_articles_all.csv", sep=';')
res_pd.fillna('NA', inplace=True)
```

```python
samples = res_pd.sample(5)
```

```python
def find_by_url_and_pick_one(url):
    tmp = ft_pd[ft_pd['<url>'] == url]
    if len(tmp.index) == 0:
        return None
    return tmp.sample(1).squeeze()
```

```python
fields = np.array(samples.iloc[1].index)[5::]

for index, sample in samples.iterrows():
    print('##############################################    {}    ##############################################'.format(index), end='\n\n')
    ft_s = find_by_url_and_pick_one(sample['<url>'])
    
    print('url:\n' + sample['<url>'], end='\n\n')
    print('title:\n' + sample['<title>'], end='\n\n')

    if ft_s is None:
        print('No record in fast track csv\n')
    else:
        for field in fields:
            print(field, end=':\n')
            print('fast track:  ' + str(ft_s[field]))
            print('our       :  ' + sample[field], end='\n\n')

    print('\n\n\n')
```
